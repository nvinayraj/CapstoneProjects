```python
import pandas as pd
import numpy as np

data = pd.read_csv("Speed_Dating_Data.csv",sep = ',',encoding = 'latin1')

#dropping columns as they are not needed for answering our questions
data["interests"] = np.nan
data.head()
```

```python
#Making all the values for the below interests as 0 for the ones that are null (or no data given). 
#The actual values would be a scale of 1 to 10, where 10 is the highest in terma of the interest shown.
data.loc[data['sports'].isnull(),['sports']] = 0
data.loc[data['tvsports'].isnull(),['tvsports']] = 0
data.loc[data['exercise'].isnull(),['exercise']] = 0
data.loc[data['museums'].isnull(),['museums']] = 0
data.loc[data['art'].isnull(),['art']] = 0
data.loc[data['hiking'].isnull(),['hiking']] = 0
data.loc[data['gaming'].isnull(),['gaming']] = 0
data.loc[data['clubbing'].isnull(),['clubbing']] = 0
data.loc[data['reading'].isnull(),['reading']] = 0
data.loc[data['tv'].isnull(),['tv']] = 0
data.loc[data['theater'].isnull(),['theater']] = 0
data.loc[data['movies'].isnull(),['movies']] = 0
data.loc[data['concerts'].isnull(),['concerts']] = 0
data.loc[data['music'].isnull(),['music']] = 0
data.loc[data['shopping'].isnull(),['shopping']] = 0
data.loc[data['yoga'].isnull(),['yoga']] = 0
data.loc[data['dining'].isnull(),['dining']] = 0
```

```python
# trying to summarize the interests into a single value where i have chosen the threshold as 6(this can be altered), 
# which means that any interest whose
# rating is given below 6 will not matter much as he/she would not spend much time on that.
interest_threshold = 6
interest_count = 0
for row in data.itertuples():
    data.loc[row.Index, 'interests'] = int(data.loc[row.Index, 'sports'] > interest_threshold) + int(data.loc[row.Index, 'tvsports'] > interest_threshold) \
    + int(data.loc[row.Index, 'exercise'] > interest_threshold) + int(data.loc[row.Index, 'dining'] > interest_threshold)\
    + int(data.loc[row.Index, 'museums'] > interest_threshold) + int(data.loc[row.Index, 'art'] > interest_threshold)\
    + int(data.loc[row.Index, 'hiking'] > interest_threshold) + int(data.loc[row.Index, 'gaming'] > interest_threshold)\
    + int(data.loc[row.Index, 'clubbing'] > interest_threshold) + int(data.loc[row.Index, 'reading'] > interest_threshold)\
    + int(data.loc[row.Index, 'tv'] > interest_threshold) + int(data.loc[row.Index, 'theater'] > interest_threshold)\
    + int(data.loc[row.Index, 'movies'] > interest_threshold) + int(data.loc[row.Index, 'concerts'] > interest_threshold)\
    + int(data.loc[row.Index, 'music'] > interest_threshold) + int(data.loc[row.Index, 'shopping'] > interest_threshold)\
    + int(data.loc[row.Index, 'yoga'] > interest_threshold)
#data.drop(columns = ['sports','tvsports', 'exercise', 'dining', 'museums', 'art', 'hiking',\
      # 'gaming', 'clubbing', 'reading', 'tv', 'theater', 'movies', 'concerts',\
      # 'music', 'shopping', 'yoga'] ,inplace=True)
data.head()
```

```python
# Dropped the below 'pid' rows for which values are not given.
data.drop(data.loc[data['pid'].isnull()].index,inplace=True)
```
